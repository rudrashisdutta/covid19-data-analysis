## Sorting the entire Dataset w.r.t. Number of Confirmed Cases in ascending order.

## Sorting the entire Dataset w.r.t. Number of Recovered Cases in descending order.

## Displaying the death and recovery percentage for all the regions.

Death Percent

```python
pd.DataFrame([data['Region'],(data['Deaths']/data['Confirmed'])*100], index=['Region', 'Death percent'])
```

Recovery Percent

```python
pd.DataFrame([data['Region'],(data['Recovered']/data['Confirmed'])*100], index=['Region', 'Recovery percent'])
```

## Displaying the Death:Recovery ratio in all regions.

```python
pd.DataFrame([data['Region'],data['Deaths']/data['Recovered']], index=['Region', 'Death:Recovery ratio'])
```
